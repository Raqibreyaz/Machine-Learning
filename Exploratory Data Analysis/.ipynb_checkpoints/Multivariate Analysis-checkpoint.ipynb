```python
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')

sns.set(style='whitegrid')
```

# Scatter Plot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')
tips
```

# Bar Plot (Numerical-categorical)

```python
sns.barplot(data=titanic,x='pclass',y='age')
```

# Box Plot (Numerical-Categorical)

```python
titanic
sns.boxplot(data=titanic,x='sex',y='age',hue='survived')
```

# Heat Map (Categorical,Categorical) 

```python
# a dataframe for the count of survived for each class
dt = pd.crosstab(titanic.pclass,titanic.survived)
sns.heatmap(dt,annot=True)
```

```python
# grouping by both sex and embark station
grouping = titanic.groupby(['sex','embarked'])
mean_grouping=grouping.mean(numeric_only=True)
(mean_grouping.survived*100).plot(kind='pie',autopct='%1.0f%%')
# for _,group in grouping:
#     print(group)
```

```python
data = flights.pivot_table(values='passengers',index='month',columns='year',observed=True)
sns.heatmap(data=data)
```

# Cluster Map (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic.parch,titanic.survived))
```

# Pair Plot

```python
sns.pairplot(iris,hue='species')
iris.head()
```

```python
sns.pairplot(data=titanic,hue='age')
```

# Line Plot

```python
sns.lineplot(data=flights.groupby('year').sum(numeric_only=True),x='year',y='passengers')
```
